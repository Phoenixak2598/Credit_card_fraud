```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```
df=pd.read_csv("https://raw.githubusercontent.com/Phoenixak2598/Credit_card_fraud/main/fraud_card.csv")
df.columns
df.shape
df1=df.sample(frac = 0.2, random_state=1)
print(df1.shape)
```

```
df1.hist(figsize = (10,10))
plt.show()
```

```
fraud = df[df['Class']==1]
non_fraud = df[df['Class']==0]
case=len(fraud)/float(len(non_fraud))
print(case)
print('Fraud Cases:{}'.format(len(fraud)))
print('Non Fraud Cases:{}'.format(len(non_fraud)))
```

```
corrmat = df1.corr()
fig = plt.figure(figsize = (20, 15))

sns.heatmap(corrmat, vmax = .6, square = True)
plt.show()
```

```
columns = df1.columns.tolist()
columns = [c for c in columns if c not in ['Class']]
target = 'Class'
X = (df1[columns]).shape
Y = (df1[target]).shape
print(X)
print(Y)
```
